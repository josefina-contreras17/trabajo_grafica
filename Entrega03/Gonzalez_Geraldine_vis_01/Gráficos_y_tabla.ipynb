```python
import pandas as pd
import altair as alt

# 1. Cargar los datos especificando el separador correcto
df = pd.read_csv('rocklatino_database_ocupada.csv.csv', sep=';')

# 2. Limpiar comillas extras en los nombres de las columnas
df.columns = df.columns.str.replace('"', '').str.strip()

# 3. Convertir la columna Year a número (por si hay errores)
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df = df.dropna(subset=['Year']) # Eliminamos filas sin año

# 4. Crear una columna para la "Década" (útil para agrupar)
df['Decade'] = (df['Year'] // 10) * 10
```

```python
# Crear el gráfico
chart = alt.Chart(df).mark_bar(cornerRadiusTopLeft=3, cornerRadiusTopRight=3).encode(
    x=alt.X('Decade:O', title='Década'), # :O porque las décadas tienen un orden
    y=alt.Y('count():Q', title='Cantidad de Canciones'), # :Q para conteo numérico
    color=alt.Color('Decade:O', scale=alt.Scale(scheme='viridis'), legend=None),
    tooltip=['Decade', 'count()']
).properties(
    title='Distribución de Canciones de Rock por Década',
    width=600,
    height=400
).interactive()

# Mostrar el gráfico
chart.show()
```
